```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.info()
```

# Univariate analysis of categorical columns(We use countplot,barplot,pieplot mostly)

```python
sns.barplot(x=df['survived'],y=df['survived'].value_counts())
#Or you can write the below code for the same 
#df['survived'].value_counts().plot(kind='bar')
df['survived'].value_counts()
```

```python
df['pclass'].value_counts().plot(kind='bar',color=['#eeae','#01ae','#11aa'])
```

```python
df['sex'].value_counts().plot(kind='bar',color=['#78ae','pink'])
```

```python
df['embarked'].value_counts().plot(kind='bar')
```

```python
df['sex'].value_counts().plot(kind='pie',autopct="%0.2f")
```

# Univariate analysis of numerical columns (Distplot,histogram,boxplot are mostly used)

```python
sns.distplot(df['age'])
```

```python
sns.boxplot(df['age'],orient="horizontal")
```
