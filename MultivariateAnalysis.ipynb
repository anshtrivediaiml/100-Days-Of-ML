```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
titanic=sns.load_dataset('titanic')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

# Numerical- Numerical

```python
plt.scatter(x=tips['total_bill'],y=tips['tip'])
#It shows that there is a linear relation between total_bill and tip , therefore when the total bill increases the tip also increases in most cases
```

```python

#Multivariate Analysis
plt.figure(figsize=(10,7))
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'])
```

# Numerical-Categorical

## barplot

```python
sns.barplot(x=titanic['pclass'],y=titanic['age'],hue=titanic['sex'])
#It can be seen from the bar chart below that the average age of pclass 1 is maximum and then pclass 2 and then pclass 3 as per sex
```

```python
sns.barplot(x=titanic['pclass'],y=titanic['fare'],hue=titanic['sex'])
# It can be seen that the avg fare of pclass 1 is max and then pclass 2 and then pclass 3 as per gender
```

## Box Plot

```python
sns.boxplot(x=titanic['sex'],y=titanic['age'],hue=titanic['survived'])
```

## Dist Plot

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
#The below graph shows the pdf of age of those peope who didnt survived and those who didnt survived 0= not survived and 1=survived
```

# Categorical-Categorical

## Heatmap

```python
grid=pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(grid)
```

```python
(titanic.groupby('sex')['survived'].mean()*100).plot(kind='bar',color=['red','green'])
```

```python
(titanic.groupby('embarked')['survived'].mean()*100).plot(kind='bar')
```

## Cluster Map

```python
var=pd.crosstab(titanic['parch'],titanic['survived'])
```

```python
sns.clustermap(var)
```

## Pair Plot

```python
sns.pairplot(iris,hue='species')
```
